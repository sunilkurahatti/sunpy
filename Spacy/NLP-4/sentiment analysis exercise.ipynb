```python
import pandas as pd
```

```python
df=pd.read_csv('moviereviews.tsv',sep='\t')
```

```python
df.head()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid=SentimentIntensityAnalyzer()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
blanks=[]
for i,lb,rv in df.itertuples():
    if type(rv)==str:
        if rv.isspace():
            blanks.append(i)
```

```python
blanks 
```

```python
df.drop(blanks)
```

```python
df['score']=df['review'].apply(lambda rv: sid.polarity_scores(rv))
```

```python
df['comp_score']=df['score'].apply(lambda d : d['compound'])
```

```python
df['comp_label']=df['comp_score'].apply(lambda sc: 'pos' if sc>=0 else 'neg')
```

```python
df.head()
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
confusion_matrix(df['comp_label'],df['label'])
```

```python
accuracy_score(df['comp_label'],df['label'])
```

```python
classification_report(df['comp_label'],df['label'])
```
