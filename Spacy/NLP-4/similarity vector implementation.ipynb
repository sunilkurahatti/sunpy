```python
from scipy import spatial
```

```python

cos_similarity=lambda vec1,vec2: 1-spatial.distance.cosine(vec1,vec2)
```

```python
king=nlp.vocab['king'].vector
man=nlp.vocab['man'].vector
queen=nlp.vocab['queen'].vector
```

```python
new_vec=king-man+queen
```

```python
similarity_vec=[]
for word in nlp.vocab:
    if word.has_vector:
        if word.lower:
            if word.is_alpha:
                similarity=cos_similarity(new_vec,word.vector)
                similarity_vec.append((word, similarity))
                
```

```python
similarity_vec=sorted(similarity_vec, key=lambda item:-item[1])
```

```python
similarity_vec
```

```python
print([t[0].text for t in similarity_vec[:20]])
```
