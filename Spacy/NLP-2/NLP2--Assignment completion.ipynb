###  For every token in the third sentence, print the token text, the POS tag, the fine-grained TAG tag, and the description of the fine-grained tag.

#### token printing format method for POS

```python
def token_print(doc):
    for token in doc:
        print('{:10s} {:10s} {:8s} {}'.format(token.text,token.pos_, token.tag_,str(spacy.explain(token.tag_))))
```

```python
count=0
for sent in doc_m.sents:
    if count !=0 and count%3==0:
        token_print(sent)
    count+=1
```

#### 4 3 . Provide a frequency list of POS tags from the entire document

```python
pos_counts=doc_m.count_by(spacy.attrs.POS)

for k,v in sorted(pos_counts.items()):
    print('{:<8}{:10s}{:8}'.format(k,doc_m.vocab[k].text,v))
```

```python
percent = 100*pos_counts[92]/len(doc_m)

print('{}/{}={}'.format(pos_counts[92],len(doc_m),percent))
```
